```python
for col in df_cleaned.select_dtypes(include='object').columns:
    unique_values=df_cleaned[col].nunique()
    if unique_values>5:
        le=LabelEncoder()
        df_cleaned[col]=le.fit_transform(df_cleaned[col])
        print(f"label encoder:{col}")
    else:
        df_cleaned=pd.get_dummies(df_cleaned,columns=[col])
        print(f"one hot encoded:{col}")
        

df_cleaned
```

```python
pps_matrix=pps.matrix(df_cleaned_1)#.sort_values(by="ppscore",ascending=False)
pps_matrix
```

```python
pps_income_all = pps_matrix[pps_matrix['y'] == 'income_ >50K']
print(pps_income_all[['x', 'ppscore']].sort_values(by='ppscore', ascending=False))
```

```python
df_cleaned_1 = df_cleaned_1.dropna(subset=['income_ >50K'])

# Ensure that the target column is numeric, if not convert it
df_cleaned_1['income_ >50K']= pd.to_numeric(df_cleaned_1['income_ >50K'], errors='coerce')

# Calculate the correlation matrix
correlation = df_cleaned_1.select_dtypes(include='number').corr()

# Sort correlation with 'income_>50K'
cor_target = correlation['income_ >50K'].sort_values(ascending=False)

# Output the result
print(cor_target)
```
